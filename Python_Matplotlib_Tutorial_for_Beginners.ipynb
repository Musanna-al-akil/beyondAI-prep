```python
import matplotlib.pyplot as plt

%matplotlib inline

from matplotlib.ticker import MaxNLocator
```

```python
apl_price = [93.95, 112.15, 104.05, 144.85, 169.49]
ms_price = [39.01, 50.29, 57.05, 69.98, 94.39]
year = [2014, 2015, 2016, 2017, 2018]
```

```python
plt.plot(year,apl_price)
plt.plot(year,ms_price)
plt.scatter(year,ms_price)
plt.xlabel('Year')
plt.ylabel("Stock Price")
plt.show()
```

```python
plt.plot(year,apl_price, ':k',
         year,ms_price)
plt.xlabel('Year')
plt.ylabel("Stock Price")

plt.axis([2013, 2019,20,180])

plt.show()
```

```python
fig_1 = plt.figure(1, figsize=(6.4, 4.8))

chart_1 = fig_1.add_subplot(221)

chart_2 = fig_1.add_subplot(222)

chart_1.plot(year,apl_price)
chart_2.scatter(year,ms_price)


plt.show()
```
